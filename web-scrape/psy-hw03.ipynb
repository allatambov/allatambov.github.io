## Домашнее задание 3

### Задание 1 (10 баллов)

Используя документацию API ВКонтакте и модуль `requests`, напишите код, который выгружает список id сообществ пользователя с id равным 171544496 и сохраняет его в переменную `group_ids`.

**Подсказка:** нужен метод из группы методов *Groups*.

### Задание 2 (5 баллов)

Напишите функцию `get_groups_info()`, которая принимает на вход числовой id сообщества, а возвращает список из следующих элементов: 

* id группы (число);
* название группы (текст);
* описание группы (текст);
* количество участников сообщества (число).

**Подсказка:** нужен метод из группы `Groups`, который начинается с `get` (выберите подходящий из перечня).

### Задание 3 (10 баллов)

Из двух вариантов решения выберите один и реализуйте:

1. Примените функцию `get_groups_info()` ко всем элементам списка `groups_ids` с задержкой по времени, сохраните полученные результаты в промежуточную структуру данных и преобразуйте её в датафрейм `final` с понятными названиями столбцов. 


2. Напишите новую модификацию функции из предыдущей задачи, которая позволяет принимать на вход список из числовых id сообществ вместо единичного id и возвращает список/кортеж с результатами. Преобразуйте полученный список в датафрейм `final` с понятными названиями столбцов.
