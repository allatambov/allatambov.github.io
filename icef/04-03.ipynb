```python
grades = {"Ann" : 5, "Claude" : 3, "Dick" : 4, 
         "Nick" : 2, "Sue" : 4} 
```

```python
excel = []  # list() 
not_excel = []

for name, grade in grades.items():
    if grade == 5:
        excel.append(name)
    else:
        not_excel.append(name) 
```

```python
excel
```

```python
not_excel
```
