```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an array with numbers from 0 to 2 with step 0.01
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
# create an array with values of a function
y = 1 + np.sin(2 * np.pi * x)
y
```

```python
# subplots returns a tuple of two objects
# figure itself and an array with its parameters
plt.subplots() 
```

```python
# save an image to fig and 
# an array with parameters to ax
fig, ax = plt.subplots()
```

```python
# create a plot
ax.plot(x, y) 
```

```python
# look at the graph
fig
```

```python
# set labels and add grid
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$")
ax.set_title("$f(x) = 1 + \sin(2 \cdot \pi \cdot x)$")
ax.grid()
```

```python
# create an array x with numbers
# f1  values of the function for the first graph
# f2  values of the function for the second one
x = np.linspace(-10, 10, 1000)
f1 = np.sin(x)
f2 = np.cos(x)
```

```python
# add two graphs to the same plot
fig, ax = plt.subplots()
ax.plot(x, f1, color = "red")
ax.plot(x, f2, color = "blue")
ax.set(xlabel='x', ylabel='y', xlim=(-9,9))
ax.grid() 
```

```python
# add labels and legend
fig, ax = plt.subplots()
ax.plot(x, f1, color = "red", linestyle='--', 
        label = "$f(x) = \sin(x)$")
ax.plot(x, f2, color = "blue", linestyle='-.', 
        label = "$g(x) = \cos(x)$")
ax.set(xlabel='x', ylabel='y',xlim=(-9, 9))
ax.legend(loc = 'upper right', framealpha = 1, 
          facecolor = "lime", title = "Legend")
ax.grid() 
```

```python
import pandas as pd
```

```python
# work with real data
df = pd.read_csv("wgi_fh_new.csv", decimal = ",")
df = df.dropna()
```

```python
df.head()
```

```python
# to get a columns with text
def get_status(x):
    if x <= 2.5:
        st = 'free'
    if x > 2.5 and x <= 5:
        st = 'partly free'
    if x > 5:
        st = 'not free'
    return st
```

```python
df['status'] = df['fh'].apply(get_status)
```

```python
# save to three different datasets
df_free = df[df.status == 'free']
df_part = df[df.status == 'partly free']
df_not = df[df.status == 'not free']
```

```python
# three graphs in one row
fig, ax = plt.subplots(nrows=1, ncols=3)
fig.set_size_inches(80, 20)
```

```python
# plot a histogram and add attributes
ax[0].hist(df_free.va, bins=5)
ax[1].hist(df_part.va, bins=5)
ax[2].hist(df_not.va, bins=5)

ax[0].set_xlim(-2.5, 2.5)
ax[1].set_xlim(-2.5, 2.5)
ax[2].set_xlim(-2.5, 2.5) 
```

```python
# the same, but in one loop
fig, ax = plt.subplots(nrows=1, ncols=3)
fig.set_size_inches(80, 20)
fig.suptitle("Voice and Accountability", fontsize = 60)

data = [df_free, df_part, df_not]
for i in range(0,3):
    ax[i].hist(data[i].va, bins=5, 
               color = "hotpink", edgecolor="navy")
    ax[i].set_xlim(-2.5, 2.5)
    ax[i].set_ylim(0, 30)
    ax[i].tick_params(labelsize = 60, length = 20)
```
